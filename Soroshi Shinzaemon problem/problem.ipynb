```python
#Problem 1
import matplotlib.pyplot as plt

total = 0
daily_rices = []
total_rices = []

for number_days in range(1, 101):
    number_rices = 2**(number_days-1)
    total += number_rices
    daily_rices.append(number_rices)
    total_rices.append(total)

print("Total number of rices by day 100: {}".format(total))

plt.figure(figsize=(12, 6))
plt.plot(range(1, 101), daily_rices, label="Rice on day n", color="blue")
plt.plot(range(1, 101), total_rices, label="Total rice by day n", color="orange")
plt.xlabel("Day")
plt.ylabel("Number of rice grains")
plt.title("Daily and total rice counts")
plt.legend()
plt.yscale("log")
plt.grid(True)
plt.show()


#Problem 2
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Sorori Shinzaemon.

    Parameters
    ----------
    n_days: int
        The number of days for which to calculate rice grains (default: 100).

    Returns
    -------
    list_n_grains : list
        A list of the number of rice grains received each day (length equals n_days).
    list_total_grains : list
        A list of the total number of rice grains accumulated by each day (length equals n_days).
    """
    list_n_grains = []
    list_total_grains = []
    total_rice = 0

    for day in range(1, n_days + 1):
        daily_rice = 2**(day - 1)
        total_rice += daily_rice
        list_n_grains.append(daily_rice)
        list_total_grains.append(total_rice)

    return list_n_grains, list_total_grains

n_days = 81
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=n_days)

print(f"Rice grains received on day {n_days}: {list_n_grains[-1]}")
print(f"Total rice grains by day {n_days}: {list_total_grains[-1]}")

plt.figure(figsize=(12, 6))
plt.plot(range(1, n_days + 1), list_n_grains, label="Rice on Day n", color="blue")
plt.plot(range(1, n_days + 1), list_total_grains, label="Total Rice by Day n", color="orange")
plt.xlabel("Day")
plt.ylabel("Number of Rice Grains")
plt.title(f"Doubling Rice Problem: Daily and Total Rice Counts (Days = {n_days})")
plt.legend()
plt.yscale("log")
plt.grid(True)
plt.show()

#Problem 3
def calculate_days_of_rice(total_rice, num_people):
    grains_per_person_per_day = 10000
    total_rice_per_day = grains_per_person_per_day * num_people
    days = total_rice // total_rice_per_day
    return days

total_rice = 2**(100)
num_people = 1000

days_lived = calculate_days_of_rice(total_rice, num_people)
print(f"With {total_rice} grains of rice, {num_people} people can live for {days_lived} days.")
```
